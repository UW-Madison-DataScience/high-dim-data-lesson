```python
# were those the best columns?
# create dummy variables?

# feature array
with pd.option_context('max_columns', None):
    cols_to_drop = ['GarageYrBlt', 'LotFrontage', 'MasVnrArea', 'SalePrice', 'hue']
    X = df[numeric_cols].drop(cols_to_drop, axis=1)
#    y = df['SalePrice'].tolist()
    y = df['top_10'].tolist()

    print(X[X.isna().any(axis=1)])
```

PCA STEP GOES HERE

```python
from sklearn.decomposition import PCA
p = PCA(n_components=2)
X_pca = p.fit_transform(X)
```

```python
# how does this line up with sales price like that? Seems like the coloration is incorrect!
fig, ax = plt.subplots(1,1, figsize=(8,6))
# top_30_idxs = [x for x in y where x==1]
scatter_plot = ax.scatter(X_pca[:,0], X_pca[:,1],
                          c=y,
                          cmap='flag',
                          alpha=0.3)
ax.set_title('scatter plot of pc1 vs pc0 shaded by SalePrice')
ax.set_xlabel('pc0')
ax.set_ylabel('pc1')
# ax.legend()
plt.show()
```

# What Just Happened !?

## Intuition: 

#### PCA is a way to rotate the *axes* of your dataset around the *data* so that the axes line up with the *directions of the greatest variation* through the data.
