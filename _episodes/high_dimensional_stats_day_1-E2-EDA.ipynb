# Create Pairplots of the top five most correlated values

```python
# Plot correlations and histograms of those columns, color with hue. This might take a while.
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(
    split_df[corr_cols + ['hue', 'SalePrice']],
    hue='hue',
    dropna=True,
    corner=True,
    plot_kws={'alpha': 0.3}
)

plt.show()
```

# Baseline - accuracy if you always predict "not top 10 %"

```python
print("{:.2f}".format(
    1 - split_df.hue.mean()
))
```

```python
# drop all *cols* containing nans (non-optimal, for demo only)
model_num_cols = split_df.describe().columns.tolist()
model_df = split_df[model_num_cols].dropna(axis=1)
model_df
```

```python
# confirm model cols, no target variable
X_cols = [x for x in model_df.columns.tolist() 
          if x not in ['SalePrice', 'hue', 'top_10']]
print(X_cols)
```

# Linear model:
## all non-null, numeric dimensions

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, log_loss
from sklearn.linear_model import LogisticRegression

scaler = preprocessing.StandardScaler().fit(model_df[X_cols])
X = scaler.transform(model_df[X_cols])
y = model_df['hue'] # 0/1 equivalent to top10 or not.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

clf = LogisticRegression().fit(X_train, y_train)
all_cols_acc = clf.score(X_test, y_test)
ll = log_loss(y_test, clf.predict(X_test))

cm = confusion_matrix(y_test, clf.predict(X_test))
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.heatmap(cm, annot=True, ax=ax, fmt='.0f')
ax.set_title('all non-nan numeric cols, linear model\n acc: {:.2f}, log loss {:.2f}'.format(all_cols_acc, ll))
ax.set_ylabel('True Label')
ax.set_xlabel('Predicted Label')
plt.show()

```

# Linear model:
## top 5 most target correlated dimensions only

```python
# define and scale features (X), and define target y
scaler_top5 = preprocessing.StandardScaler().fit(model_df[corr_cols])
X_top5 = scaler_top5.transform(model_df[corr_cols])
y_top5 = model_df['hue'] # 0/1 equivalent to top10 or not.

# split data
X_train_top5, X_test_top5, y_train_top5, y_test_top5 = train_test_split(X_top5, y_top5, test_size=0.33)

# create classifier
clf_top5 = LogisticRegression().fit(X_train_top5, y_train_top5)

# assess classifier
all_cols_acc_top5 = clf_top5.score(X_test_top5, y_test_top5)
ll_top5 = log_loss(y_test_top5, clf_top5.predict(X_test_top5))
cm_top5 = confusion_matrix(y_test_top5, clf_top5.predict(X_test_top5))

# plot chart
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.heatmap(cm_top5, annot=True, ax=ax, fmt='.0f')
ax.set_title('all non-nan numeric cols, linear model\n acc: {:.2f}, log loss {:.2f}'.format(all_cols_acc_top5, ll_top5))
ax.set_ylabel('True Label')
ax.set_xlabel('Predicted Label')
plt.show()
```

# Linear Model:
## 2D PCA of all columns

```python
from sklearn.decomposition import PCA

p = PCA(n_components=2)
X_pca = p.fit_transform(model_df[X_cols])

y_pca = model_df['hue']
```

```python
X_pca
```

```python
# split data
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_pca, test_size=0.33)

# create classifier
clf_pca = LogisticRegression().fit(X_train_pca, y_train_pca)

# assess classifier
all_cols_acc_pca = clf_pca.score(X_test_pca, y_test_pca)
ll_pca = log_loss(y_test_pca, clf_pca.predict(X_test_pca))
cm_pca = confusion_matrix(y_test_pca, clf_pca.predict(X_test_pca))

# plot results
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.heatmap(cm_pca, annot=True, ax=ax, fmt='.0f')
ax.set_title('all non-nan numeric cols, linear model\n acc: {:.2f}, log loss {:.2f}'.format(all_cols_acc_pca, ll_pca))
ax.set_ylabel('True Label')
ax.set_xlabel('Predicted Label')
plt.show()
```

```python
# create df from pca array and add in category information
X_pca_df = pd.DataFrame(X_pca)
X_pca_df['category'] = model_df['hue'].reset_index(drop=True)
X_pca_df
```
